```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('/content/data_penjualan.csv', delimiter=';') # Use semicolon as delimiter

# Membuat visualisasi grafik garis
plt.figure(figsize=(8, 5))
plt.plot(data['Bulan'], data['Penjualan'], marker='o', linestyle='-', color='b')

# Menambahkan judul dan label
plt.title('Penjualan per Bulan', fontsize=14)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Penjualan', fontsize=12)

# Menampilkan grid dan grafik
plt.grid(True)
plt.tight_layout()
plt.show()
```
